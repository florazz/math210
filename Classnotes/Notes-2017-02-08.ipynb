## 2. Plotting in 3D with mplot3d

```python
from mpl_toolkits.mplot3d import Axes3D 
```

```python
plt.figure() # Create a new figure
ax= plt.axes(projection='3d') # Create a new Axes3D object in the current figure

x=np.linspace(-1,1,100)
y=np.linspace(-1,1,100)
X,Y = np.meshgrid(x,y)
Z=1-X**2-Y**2
ax.plot_surface(X,Y,Z)
```

Let's plot $z=\cos(\sqrt{x^2+y^2})$ for $x,y\in [-4\pi,4\pi]$:

```python
plt.figure() # Create a new figure
ax= plt.axes(projection='3d') # Create a new Axes3D object in the current figure

x=np.linspace(-8*np.pi,8*np.pi,500)
y=np.linspace(-8*np.pi,8*np.pi,500)
X,Y = np.meshgrid(x,y)
Z=np.cos(np.sqrt(X**2+Y**2))
ax.plot_surface(X,Y,Z,cmap=plt.cm.spring),ax.set_zlim(bottom=-5,top=5)
```

### What is meshgrid?

Like many 3D plotting libraries (ie.MatLab), the 3D plotting functions require 2D matrices of x,y and z values. Meshgrid takes to arrays and returns two 2D matices X and Y with X has the x values across its rows and Y has the y values down its columns.

```python
x=[1,2,3]
y=[8,9,10,11]
```

```python
np.meshgrid(x,y)
```

The output of meshgrid is a list with 2 entries. We can unpack those entries into variables X and Y.

```python
X,Y =np.meshgrid(x,y)
```

Let's plot a sphere! A 3D parametric plot with 2 parameters $\theta$ and $\phi$ (in spherical coordinates).

```python
plt.figure() # Create a new figure
ax= plt.axes(projection='3d') # Create a new Axes3D object in the current figure

phi=np.linspace(0,np.pi,50)
theta=np.linspace(0,2*np.pi,50)

PHI, THETA = np.meshgrid(phi,theta)

X= np.sin(PHI)*np.cos(THETA)
Y= np.sin(PHI)*np.sin(THETA)
Z=np.cos(PHI)

ax.plot_wireframe(X,Y,Z,cmap=plt.cm.summer)
plt.savefig('sphere.png',format='png')
```
