## 1. More plotting examples

### Example: Trefoil knot in 2D

The trefoil knot in 2D is given by the parametric equations:

$$
x(t) = sin(t)+ 2sin(2t),
y(t) = cos(t) - 2cos(2t),
t\in [0,2\pi]
$$

```python
t=np.linspace(0,2*np.pi,1000)
x=np.sin(t)+2*np.sin(2*t)
y=np.cos(t)-2*np.cos(2*t)
```

```python
plt.plot(x,y)
plt.axis('equal')
plt.xlim([-4,4])
plt.ylim([-4,3])
plt.title('Trefoil knot in 2D')
```

### Example: Plot several curves together

Plot the curve $y= e^{-x^2}cos(kx)$ for $k=1,2,5,10$ for $x\in [-5,5]$.

```python
x = np.linspace(-2,2,1000)
ks = np.random.rand(100)*10
for k in ks:
    y=np.exp(-x**2)*np.cos(k*x)
    plt.plot(x,y)
```

## 2. Array methods

NumPy arrays have many methods (ie. functions) to do transformations and calculations. We access methods using the dot rotation following the array variable name. For example, there is the `sum` method which simply sums the entries in the array.

```python
arr = np.array([1,2,3,4,5])
```

```python
arr.sum()
```

There is an equivalent functional way of doing the same thing:

```python
np.sum(arr)
```

There is also the prod method:

```python
arr.prod()
```

The method `cumsum` computes the cumulative sum of the entries:

```python
arr.cumsum()
```

For example, we can plot a random walk:

```python
x = np.random.randn(1000,5).cumsum(axis=0)
plt.plot(x)
```

We can access the shape of an array using `.shape`

```python
M = np.random.randint(0,2,[4,5])
print(M)
M.shape
```

Notice that we accessed the shape without using parentheses. This is the difference between array methods (functions to do calculations with the array) and array attributes (values associated with the array - like its shape).

```python
M.sum()
```

```python
M.shape
```

### 3. Array slicing and indexing

We access the entries in an array using the square brackets:

```python
A= np.random.randint(0,10,[3,5])
A
```

```python
A[2,4] # Row index 2 and column index 4
```
